```python
######! error ?????

def remove_outlier(df):
    Q1 = df.quantile(.25)
    Q3 = df.quantile(.75)
    iqr = Q3-Q1
    df[(df < (Q1-1.5*iqr)) or (df > (Q3+1.5*iqr))] = np.nan
    df.dropna()
    return df

df = remove_outlier(df)
df.head()
```
